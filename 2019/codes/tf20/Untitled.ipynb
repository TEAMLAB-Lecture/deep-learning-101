```python
@tf.function
def add(a,b):
    return a + b

add
```

```python
add(tf.ones([2,2]), tf.ones([2,2]))
```

```python
a = tf.Variable(1.0)
b = tf.Variable(2.0)

@tf.function
def add(x,y):
    a.assign(y*b)
    b.assign_add(x*a)
    print(a,b)
    return a + b

add(1.0,2.0)
```

```python
@tf.function
def f(x):
    while tf.reduce_sum(x) > 1:
        x=tf.tanh(x)
    return x

f(tf.random.uniform([10]))
```
